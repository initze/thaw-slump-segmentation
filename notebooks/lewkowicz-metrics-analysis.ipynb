```python
import pandas as pd
import seaborn as sns

# Set a nice color palette
palette = sns.color_palette('husl', 8)
sns.set_theme(palette=palette, context='notebook', style='whitegrid')
```

```python
metrics = pd.read_parquet('data/publication/v5/lewkowicz_metrics.parquet')
metrics_max_dissolve = metrics[metrics['year'] == 0]
metrics = metrics[metrics['year'] != 0]
metrics
```

```python
g = sns.FacetGrid(metrics, col='region', row='metric', hue='year', sharey='row', sharex=False, margin_titles=True)
g.map(sns.lineplot, 'buffer', 'value')
g.add_legend()
```

```python
g = sns.FacetGrid(metrics, col='metric', hue='year', sharey=False, sharex=False, margin_titles=True)
g.map(sns.lineplot, 'buffer', 'value')
g.add_legend()
```

```python
g = sns.FacetGrid(metrics, col='metric', sharey=False, sharex=False, margin_titles=True)
g.map(sns.lineplot, 'buffer', 'value')
g.add_legend()
```

```python
g = sns.FacetGrid(metrics_max_dissolve, col='metric', hue='region', sharey=False, sharex=False, margin_titles=True)
g.map(sns.lineplot, 'buffer', 'value')
g.add_legend()
```

```python
g = sns.FacetGrid(metrics_max_dissolve, col='metric', sharey=False, sharex=False, margin_titles=True)
g.map(sns.lineplot, 'buffer', 'value')
g.add_legend()
```
